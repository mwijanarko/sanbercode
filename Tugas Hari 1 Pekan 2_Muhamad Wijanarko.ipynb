<h2>Soal 5: Deskriptif Statistik</h2>

Jelaskan apa itu deskriptis statistik!

Jawab: 
Statistik deskriptif adalah metode yang cukup sederhana dari statistika dimana operasi yang dilakukan terhadap suatu data adalah pengorganisasian data, penjumlahan (rata-rata, median, dst), serta teknik penyajian yang mudah dipahami. Pembatasannya adalah statistik deskriptif tidak memiliki kemampuan untuk dijadikan alat penarik kesimpulan, tetapi cukup untuk menggambarkan kondisi suatu data yang sedang diteliti.

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here

x = df_from_csv.describe()
print(x)

y = df_from_csv['price'].mean()
print("Rata-rata dari Price adalah: ",y)

z = df_from_csv['depth'].std()
print("Standar deviasinya adalah ",z)

max_carat = df_from_csv['carat'].max
print(max_carat)
```
